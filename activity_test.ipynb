# 1. Model Training

```python
import pandas as pd
```

```python
x_train = pd.read_csv('./x_train.csv', index_col=0)
y_train = pd.read_csv('./y_train.csv', index_col=0)
x_test = pd.read_csv('./x_test.csv', index_col=0)
y_test = pd.read_csv('./y_test.csv', index_col=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

rf = RandomForestClassifier(n_estimators=500)
rf.fit(x_train, y_train)
```

```python
pred = rf.predict(x_test)
accuracy = accuracy_score(y_test, pred)

print(f'Mean accuracy score: {accuracy:.5}')
```

# 2. Model Testing

```python
import os
import pandas as pd

# Open test data file
dir = './mlpr20_project_test_data/'
list_files = os.listdir(dir)
file = open(dir+list_files[0])

columns = ['acc_x', 'acc_y', 'acc_z', 'ang_x', 'ang_y', 'ang_z']
test_x = pd.read_csv(file, names=columns, sep='\s+')
```

```python
pred = rf.predict(test_x)
```

```python
i=0

f = open('result_rf.txt', 'w')

for x in range(len(list_files)):
    count_rf = open(dir+list_files[i])
    test_x = pd.read_csv(count_rf, names=columns, sep='\s+')
    pred = rf.predict(test_x)
    count = collections.Counter(pred)
    p = count.most_common()[0][0]
    p = int(p)
    f.write(str(p))
    f.write('\n')
    i=i+1
        
f.close()
```
